```python
df_match = pd.DataFrame(data[0]['matchSummary'])
df_match.head()
```

```python
# make scorecard as primary key so that we can link to many tables
df_match.rename({'scorecard':'match_id'},axis = 1, inplace = True)
df_match.head()
```

```python
df_match.shape
```

```python
match_ids_dict = {}
for index,row in df_match.iterrows():
    key1 =  row["team1"]+" Vs "+row["team2"]
    key2 = row["team2"]+" Vs " + row["team1"]
    match_ids_dict[key1] = row["match_id"]
    match_ids_dict[key2] = row["match_id"]
match_ids_dict
```

**batting_summary**

```python
with open("C:\\Users\\Bhavya Sree\\Downloads\\t20_wc_batting_summary.json") as f:
    data = json.load(f)
    all_records = []
    for rec in data:
        all_records.extend(rec["battingSummary"])
df_batting = pd.DataFrame(all_records)
df_batting.head(10)
```

```python
df_batting.shape
```

```python
df_batting["out/notout"] = df_batting.dismissal.apply(lambda x: "out" if len(x) > 0 else "notout")
df_batting
```

```python
df_batting.drop(columns=['dismissal'],inplace = True)
df_batting.head(10)
```

```python
df_batting["batsmanName"] = df_batting["batsmanName"].apply(lambda x:x.replace("â€",""))
df_batting["batsmanName"] = df_batting["batsmanName"].apply(lambda x:x.replace("\xa0",""))
df_batting
```

```python
df_batting.head(40)
```

```python
df_batting["match_id"] = df_batting["match"].map(match_ids_dict)
df_batting.head(40)
```

```python
df_batting[df_batting["teamInnings"]=="India"]
```

```python
import os

directory = 't20_csv_files'
if not os.path.exists(directory):
    os.makedirs(directory)

df_batting.to_csv(os.path.join(directory, 'temp.csv'))
```
